```python
os.makedirs("../Data/GDSC/Drugs/", exist_ok=True)
os.makedirs("../Results/GDSC/Drug Plots/", exist_ok=True)
data = pd.read_csv("../Data/GDSC/IC50s.csv").drop(["Unnamed: 0"], axis=1)
drug_names = data["DRUG_NAME"].value_counts().index.tolist()
drug_names = sorted(drug_names)
for drug_name in drug_names:
    if drug_name == "UNC1215":
        drug_data = data[data["DRUG_NAME"] == drug_name].drop(["DRUG_NAME"], axis=1)
        max_micromolar = drug_data["MAX_CONC_MICROMOLAR"].iloc[0]
        min_micromolar = drug_data["MIN_CONC_MICROMOLAR"].iloc[0]
        threshold = (max_micromolar + min_micromolar) / 2.0
        print(drug_data["LN_IC50"].min(), math.log(threshold), drug_data["LN_IC50"].max(), end="\t")
        res_indices = drug_data.loc[data["LN_IC50"] >= (math.log(threshold))].index # resistant
        sen_indices = drug_data.loc[data["LN_IC50"] < (math.log(threshold))].index # sensitive
        drug_data.loc[res_indices, "Label"] = 1
        drug_data.loc[sen_indices, "Label"] = 0
        drug_data.drop(["MIN_CONC_MICROMOLAR", "MAX_CONC_MICROMOLAR"], axis=1)
        drug_name = "-".join(drug_name.split("/"))
        drug_data.to_csv("../Data/GDSC/Drugs/" + drug_name + ".csv", index=None)
        plt.close("all")
        plt.figure(figsize=(15, 10))
        sns.distplot(drug_data["LN_IC50"])
        plt.axvline(x=math.log(threshold), color='r', linestyle='dashed')
        plt.ylabel("Frequency")
        plt.xlabel("Log(IC50)")
        plt.title("%s Distibution (Threshold = %.4f)" % (drug_name, math.log(threshold)))
        plt.savefig("../Results/GDSC/Drug Plots/" + drug_name + ".pdf")
    #     print(len(drug_data[drug_data["LN_IC50"] > math.log(threshold)]))
    #     print(len(drug_data[drug_data["LN_IC50"] <= math.log(threshold)]))
        print(drug_name, "Is Finished!")
        print(drug_data)
```
