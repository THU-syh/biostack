# Step 2 - Read CYP Drug Molecular Descriptors

```python
cyp_drug_mds = pd.read_csv('CYP_Drugs_Molecular_Descriptors_NUMERIC_FORMAT.tab',sep='\t')
```

```python
# Search by Name
supercyp_names = list(set(list(supercyp['Name'])))
```

```python
supercyp_names = supercyp_names[1:] # Remove nan
supercyp_names
```

```python
supercyp_correct_names = []
for name in supercyp_names:
    try:
        n = str(name)
        if n[-1] == ' ':
            n = n[0:len(n)-1]
            supercyp_correct_names.append(n.lower())
            continue
        else:
            supercyp_correct_names.append(n.lower())
            continue
    except:
        n = str(name[0:len(name)-1])
        supercyp_correct_names.append(n.lower())
```

# Merge both dataframes

```python
supercyp_identifiers = []
for name in list(supercyp['Name']):
    try:
        n = str(name)
        if n[-1] == ' ':
            n = n[0:len(n)-1]
            supercyp_identifiers.append(n.lower())
            continue
        else:
            supercyp_identifiers.append(n.lower())
            continue
    except:
        n = str(name[0:len(name)-1])
        supercyp_identifiers.append(n.lower())
supercyp['Drug Identifier'] = supercyp_identifiers
```

```python
del supercyp['CAS RN']
del supercyp['Name']
del supercyp['Relation']
del supercyp['Species']
```

```python
df = pd.merge(supercyp,cyp_drug_mds,on='Drug Identifier')
df.head()
```
