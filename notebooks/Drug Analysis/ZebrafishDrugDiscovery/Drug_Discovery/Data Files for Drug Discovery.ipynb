```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt
import numpy as np

cap = cv2.VideoCapture("20190703.164704.S11-60s.mkv")
# pictures = []
# prev = 0
i=0
# prev = [][]
x = 1
motion = []
MI = []
while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()
    if(frame is None):
        break
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY).astype(np.float64)
    gray = np.array(gray)
    if i > 0:
        motion = np.abs(gray - prev)# > 10
        MI.append(np.sum(motion))
    prev = gray        
    i += 1
```

```python
plt.plot(MI)
plt.ylabel('Motion Index')
plt.xlabel('Frame')
plt.show()
```

```python
d = []
for i in range(len(MI)):
    if MI[i] > 2000000:
        d.append(i)
print(d)
print()
w = []
for i in range(1,len(d)):
    w.append(d[i] - d[i-1])
print(w)
```
