```python
def fix_counts(counts):
    column = counts.columns[0]
    return pd.DataFrame(counts[column])\
        .rename(columns={column: 'count'}).reset_index()

targets = biomol

targ_counts = fix_counts(targets.groupby(['ref_name', 'object_name']).count())
targ_counts
```

```python
import seaborn as sns
figure = plt.figure(figsize=(6, 4))
ax = figure.add_subplot(1,1,1)
sns.distplot(targ_counts['count'], bins=140, kde=False, rug=True, rug_kws={'color': 'black', 'alpha': 0.5}, color='blue', ax=ax)
# TODO set alpha
ax.set_xlim(0, )
```

```python
targ_counts = targ_counts.sort_values('count', ascending=False)
filtered_targ_counts = targ_counts[targ_counts['count']>2]
filtered_targ_counts2 = filtered_targ_counts.iloc[:8, :] # 61 targets
filtered_targ_counts2
```

```python
df = pd.DataFrame({'Name':filtered_targ_counts2["object_name"], 'Count':filtered_targ_counts2["count"]})
#figure = plt.figure(figsize=(8, 10))
ax = df.plot.bar(x='Name', y='Count', rot=40)
```
